Data Preparation:

```python
import pandas as pd
df=pd.read_excel("transactions.xlsx")
df
```

```python
#Output Value
y=df['Suspicious Value']
y
```

```python
#Dropping of Transaction Id and Suspicious value from training data
x=df.drop(columns=['Transaction ID','Suspicious Value'])
x
```

Converting String-data columns to numeric values

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
categorical_columns = ['From Account Name', 'From Country', 'To Account Name', 'To Country', 'Mode of Transaction', 'Risk']

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    x[column] = label_encoders[column].fit_transform(x[column])
x
```

Data Splitting      

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=100)
```

```python
x_train
```

```python
x_test
```

Model Building

Decission Tree model

```python
from sklearn.tree import DecisionTreeClassifier

dt_model= DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
```

Testing the model

```python
y_dt_model_train_pred=dt_model.predict(x_train)
y_dt_model_test_pred=dt_model.predict(x_test)
```

```python
y_dt_model_test_pred
```

Accuracy

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

accuracy = accuracy_score(y_test, y_dt_model_test_pred)
print(f"Accuracy: {accuracy:.2f}")
Randomforest_report = classification_report(y_test,y_dt_model_test_pred)
print(Randomforest_report)
```

saving the trained model using joblib

```python
import joblib
model_filename = 'DecisionTree_model.pkl'
joblib.dump(dt_model, model_filename)
model_filename  # Return the filename where the model is saved
```

RandomForest model

```python
from sklearn.ensemble import RandomForestClassifier

rf_model= RandomForestClassifier()
rf_model.fit(x_train,y_train)
```

Testing the model

```python
y_rf_model_train_pred=rf_model.predict(x_train)
y_rf_model_test_pred=rf_model.predict(x_test)
```

```python
y_rf_model_test_pred
```

Accuracy and classification Report:

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
accuracy = accuracy_score(y_test, y_rf_model_test_pred)
print(f"Accuracy: {accuracy:.2f}")

Randomforest_report = classification_report(y_test,y_rf_model_test_pred)
print(Randomforest_report)
```

saving the trained model using joblib

```python
import joblib
model_filename = 'Randomforest_model.pkl'
joblib.dump(rf_model, model_filename)
model_filename  # Return the filename where the model is saved
```

Testing out with new Data
